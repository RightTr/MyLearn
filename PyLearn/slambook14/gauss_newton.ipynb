```python
import torch
import time
import torch.autograd.functional as F
```

```python
def func(x, p):
    x = x.unsqueeze(0)
    i = torch.ones(x.shape)
    x_2 = x**2
    X = torch.cat((x_2, x, i), dim = 0).T
    n = torch.normal(0, 1, size=(100, 1))
    y = torch.exp(torch.matmul(X, p)) + n
    return X, y
```

```python
def error(x, p, p_):
    _, y = func(x, p)
    _, y_ = func(x, p_)
    return y - y_
```

```python
x = torch.arange(100) / 100
p = torch.tensor([1, 2, 1]).float().unsqueeze(1)
p_ = torch.tensor([1, -2, 5]).float().unsqueeze(1)
X, y = func(x, p)
```

```python
iterations = 100
lr = 0.1
e_sum = 0
start_time = time.time()
for i in range(iterations):
    X_, y_ = func(x, p_)
    e = y - y_
    e_sum = torch.sum(e)
    J = -(X * y_.repeat(1, 3))
    H = J.T @ J
    g = -J.T @ e
    dp = torch.inverse(H) @ g
    p_ += dp * lr
    print(f'cost: {e_sum}')
    if(dp.norm() < 1e-6):
        print(i)
        break
end_time = time.time()

print(f'result: {p_}')
print(f'duration: {end_time - start_time}')
    
```

```python
iterations = 100
lr = 0.1
e_sum = 0
start_time = time.time()
for i in range(iterations):
    e = error(x, p, p_)
    e_sum = torch.sum(e)
    J = F.jacobian(lambda p_: error(x, p, p_), p_)
    J = J.reshape(100, 3)
    H = J.T @ J
    g = -J.T @ e
    dp = torch.inverse(H) @ g
    p_ += dp * lr
    print(f'cost: {e_sum}')
    if(dp.norm() < 1e-6):
        print(i)
        break
end_time = time.time()

print(f'result: {p_}')
print(f'duration: {end_time - start_time}')
```
